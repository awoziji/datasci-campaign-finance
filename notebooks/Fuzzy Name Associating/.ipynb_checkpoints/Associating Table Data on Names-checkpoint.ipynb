# Fuzzy Name Associating

## Noteable value
clean(name)
* Converts a raw candidate name into a uniform format of 'titles Fname mi Lname'

fuzzy_match(...)
* Generates 'raw db name' to ID pairings
* Matches names against names already in this 'matched db'
    * If the match percent is bigger than or equal to the threshold then the two names are considered a match and the new name is given the same ID
    * Other wise the new name is given a new ID
    * Then makes appropriate entries into the 'matched db' aka db_dbname_2_id and db_id_2_showname

apply_add_id_column(...)
* Modifies a DataFrame to include the generated ID's from the raw db name of each row

### Helper Functions
#### clean(name)
Converts a name from the database into a standard format. 
This is necessary because the names in one of the databases are not all in a standard format. In particular there are multiple entries for similar names where titles are permuted. E.g. "BACA. JR., JOE", "BACA JR., JOE", "Joe Baca, Jr." and "MC CARTHY, KEVIN", "Kevin McCarthy*".

clean() handles: commas, middle initial, sr, jr, ii, iii, and concatenating mc to the lastname if it already prefixes the lastname.
It moves all sr, jr, ii, iii to the front of the name, and the middle initial to just before the last name.

Converts a database name into a more uniform format.
* BACA. JR., JOE --> jr joe baca
* BACA JR., JOE  --> jr joe baca
* Joe Baca, Jr.  --> jr joe baca
* MC CARTHY, KEVIN --> kevin mccarthy
* Kevin McCarthy*  --> kevin mccarthy
* MONVILLE III, LOU --> iii lou monville

#### lastname(name)
Extracts the last name from a cleaned name. 
Note: it was written before clean name was extended so it superfluously also handles mc

```python
# Helper functions

regex = re.compile(r'[^a-z ]') # Used for lower and remove none a-z characters
def clean(name):
    mi = []
    sr = []
    jr = []
    ii = []
    iii = []
    
    # Handle comma
    parts = name.split(",")
    if len(parts) == 2: # If they aren't equal we have more or less than 1 comma, which we don't know how to handle. eg not "Smith, John M"
        name = parts[1].strip() + " " + parts[0].strip()
    
    # Handle symboles
    name = regex.sub("", name.lower()) # lower and remove none a-z characters
    
    # Handle sr jr ii iii to front of the name and middle initial to the end
    parts = name.split(" ")
    i = len(parts)
    while i > 0: 
        i -= 1
        if len(parts[i]) == 1: ## Middle Initial
            mi += [parts[i]]
            del parts[i]
            continue
        if parts[i] == "sr":
            sr += [parts[i]]
            del parts[i]
            continue
        if parts[i] == "jr":
            jr += [parts[i]]
            del parts[i]
            continue
        if parts[i] == "ii":
            ii += [parts[i]]
            del parts[i]
            continue
        if parts[i] == "iii":
            iii += [parts[i]]
            del parts[i]
            continue
            
    # Handle mc lastname -> mclastname
    if len(parts) > 1 and parts[-2] == "mc":
        if not parts[-1].startswith("mc"):
            parts[-1] = "mc" + parts[-1]
            del parts[-2]
            
    # Put mi before the last name -- I noticed this causes pyjarowinkler to generate less 97%'s when two names have different mi
    parts.insert(-1, ' '.join(mi))
    
    # recombine name
    name = " ".join(sr + jr + ii + iii + parts)
    
    return name

# Take in a cleaned name and pull out the last name
#  take in a cleaned name
#  start at the last token
#  take steps towards the front every time its a ["jr", "sr"] or ["ii", "iii"] or len(part[i]) == 1 (assuming its an initial)
#  now we should be at the last name
#  if the word before this word is mc, prepend mc to this word (my tests have shown we don't also need to test that this word does not already begin with mc, but its a simple code to guard against)
#  return lastname
not_last_names = ["jr", "sr", "ii", "iii"]
def lastname(name):
    parts = name.split(" ")
    
    i = len(parts) - 1
    while len(parts[i]) == 1 or parts[i] in not_last_names:
        i -= 1
        
    if parts[i - 1] == "mc":
        if not parts[i].startswith("mc"):
            parts[i] = "mc" + parts[i]
            
    return parts[i]
```

# Fuzzy Name Associating
### Custom fuzzy name matching functions
You can make your own fuzzy name matching function using any new library you like. 
It must produce a percent as an integer from 0-100 representing how similar the two names are.
#### Examples:
* fuzz_ratio(name1, name2)
* jarowinkler(name1, name2) (default matcher)

### Fuzzy_match(sql_names, fuzzy_match_fun, match_threshold, column_name)
Give it 
* sql_results that have a "dbname" column (or pass the column name via function argument)
* a function that preforms name matching (takes in 2 names, return 1 number, percent, 0-100, int. See example fuzz_ratio() and jarowinkler())
* a match threshold to determine what constitutes a match (anything >= match_threshold is a match, defaults 100). Look at the text dump to figure out where this should be.
* (optional) a column_name of the sql column storing the name to match on. Defaults to "dbname"

Receive
* db_dbname_2_id - Pairings of raw db names to ID
* db_id_2_showname - Pairs of ID to display names (defaults to first added name)
* hits - formated as [[percent_match, "% index Hit: name1 vs name2"], [...], ...]
* misses - e.g. [[70, "% 1280 Miss: Sherry Walker vs Jerry Kowal"], ...]

hits and misses note:
* print these using print_list_of_lists(hits)
* plot these using plot_scores(hits, misses)

Note: this function has code to take human input to allow for a more nuanced approach to the grey area of matches, but was not fully ultilized and tested. (see locations of upper_limit, lower_limit, and match_threshold variables)

### Helper functions
#### print_list_of_lists(list_of_lists)
Used to print data in lists used by this function. (hits, misses, etc.)

#### plot_scores(hits, misses)
Uses Pandas DataFrame plot function to plot the frequency of percents in hits and misses

```python
# Perform matching

# Take in 2 names, return 1 number, percent, 0-100, int
# This method we try to use a fuzzywuzzy ratio of the whole_name * last_name
def fuzz_ratio(name1, name2):
    name1 = clean(name1)
    name2 = clean(name2)
    whole_name_match_ratio = fuzz.ratio(name1, name2) / 100
    last_name_match_ratio = fuzz.ratio(lastname(name1), lastname(name1)) / 100
    return int(whole_name_match_ratio * last_name_match_ratio * 100)

# Generic pyjaro implementation
def jarowinkler(name1, name2):
    name1 = clean(name1)
    name2 = clean(name2)
    whole_name_match_ratio = distance.get_jaro_distance(name1, name2)
    return int(whole_name_match_ratio * 100)
    

def fuzzy_match(sql_names, fuzzy_match_fun = jarowinkler, match_threshold = 100, column_name = "dbname"):
    
    time_start = time.time()

    UPPER_LIMIT = match_threshold # Anything >= this score will be considered "a definite match"
    LOWER_LIMIT = match_threshold # Anything <= this score will be considered "a definite miss"

    db_dbname_2_id = pd.DataFrame(columns=["dbname", "id"])  # [] # (dbname, id)
    db_id_2_showname = pd.DataFrame(columns=["id", "showname"]) # [] # (id, showname)

    # Static insert for "None" as -1 
    # db_dbname_2_id = db_dbname_2_id.append({"dbname" : "None", "id" : -1}, ignore_index=True)
    # db_id_2_showname = db_id_2_showname.append({"id" : -1, "showname" : "None"}, ignore_index=True)

    next_id = 1 # At some point this should be set equal to "retreive last id in db"
    total = len(sql_names)
    count_times = 0

    hits = []
    misses = []
    skips = []
    news = []
    nones = []
    
    # Loop over all the names
    for index, input_dbname in sql_names.iterrows():
        count_times += 1
        
        # Running percent complete ticker
        print(str(count_times) + " " + str(count_times / total * 100) + "%" + " "*10, end="\r") 
        # \r causes it to print over itself, so instead of being spammed with text, its a nice little ticker

        # Extract the name
        dbname = input_dbname[column_name]

        # Note and skip None
        if dbname is None:
            nones += [[index]]
            continue

        # Handle already in db
        if sum(db_dbname_2_id["dbname"] == dbname) > 0: # If the name is already processed
            skips += [[count_times, " Skip: " + str(dbname)]]
            continue # Skip it
        
        # Perform fuzzy matching against all names already in the db (something like O(n log n))
        fuzzymatches = []
        for index, row in db_dbname_2_id.iterrows(): # iterrows is really slow I think I've read?
            name = row["dbname"]
            id_ = row["id"] 
            fuzzymatches += [[fuzzy_match_fun(name, dbname), id_, name]]
        
        # Handle new name, pretty much only the first iteration
        if not fuzzymatches:
            news += [[count_times, " New: " + str(dbname)]]
            db_dbname_2_id = db_dbname_2_id.append({"dbname" : dbname, "id" : next_id}, ignore_index=True)
            db_id_2_showname = db_id_2_showname.append({"id" : next_id, "showname" : dbname}, ignore_index=True)
            next_id += 1
            continue # NEXT!
        
        # Sort matches, now the biggest match is at the top, consider it the best match. 
        # Note: assuming the highest match to be the ONLY __relevant__/**important** match will probably byte me in the butt later
        fuzzymatches.sort(reverse=True, key=lambda x: x[0])
        h_match_name = fuzzymatches[0][2]
        h_match_id = fuzzymatches[0][1]
        h_match_score = fuzzymatches[0][0]
        
        # Add the match
        if h_match_score >= UPPER_LIMIT:
            db_dbname_2_id = db_dbname_2_id.append({"dbname" : dbname, "id" : h_match_id}, ignore_index=True)   
            hits += [[h_match_score, "% " + str(count_times) + " Match: " + str(dbname) + " vs " + str(h_match_name)]]
        
        # Add the miss
        elif h_match_score <= LOWER_LIMIT:
            db_dbname_2_id = db_dbname_2_id.append({"dbname" : dbname, "id" : next_id}, ignore_index=True)
            db_id_2_showname = db_id_2_showname.append({"id" : next_id, "showname" : dbname}, ignore_index=True)
            next_id += 1
            misses += [[h_match_score, "% " + str(count_times) + " Miss: " + str(dbname) + " vs " + str(h_match_name)]]

        # Let a Human handle it. Note: Untested
        else:
            print(h_match_score, "% match: ", dbname, " vs ", name, " ")
            print("Same person (y/N)? ")
            answer = input()
            if answer == "":
                answer = "n"

            if answer.lower() == "y":
                db_dbname_2_id = db_dbname_2_id.append({"dbname" : dbname, "id" : h_match_id}, ignore_index=True)
            elif answer.lower() == "n":
                db_dbname_2_id = db_dbname_2_id.append({"dbname" : dbname, "id" : next_id}, ignore_index=True)
                db_id_2_showname = db_id_2_showname.append({"id" : next_id, "showname" : dbname}, ignore_index=True)
                next_id += 1

    # Finally... I mean, clear the ticker
    print(" " * 80)

    db_dbname_2_id = db_dbname_2_id.set_index("dbname") # Now we can do csv_data.loc[str(x)].id to get the id
    db_id_2_showname = db_id_2_showname.set_index("id")
    
    # Sort the complicated lists
    misses.sort(reverse=True, key=lambda x: x[0])
    hits.sort(reverse=True, key=lambda x: x[0])

    #print("Misses:")
    #print_list_of_lists(misses)
    #print("Skips:")
    # print_list_of_lists(skips)
    #print("New:")
    #print_list_of_lists(news)
    #print("Hits:")
    #print_list_of_lists(hits)
    #print("Nones indexes:")
    #print_list_of_lists(nones)

    #print(db_dbname_2_id)
    #print(db_id_2_showname)
    
    # Display statistics
    print("News:", len(news))
    print("Hits:", len(hits))
    print("Misses:", len(misses))
    print("Skips:", len(skips))
    print("Nones:", len(nones))
    print("Total:", len(hits) + len(misses) + len(skips) + len(news) + len(nones))
    print("Dataset len:", total)
    print("Result len:", len(db_dbname_2_id))

    print("Accuracy:", len(hits) / total)

    time_elapsed = time.time() - time_start
    print("Took:", time_elapsed, " seconds")
    
    return db_dbname_2_id, db_id_2_showname, hits, misses
          
def print_list_of_lists(l):
    r = []
    for i in l:
        r += [''.join(map(str,i))]
    print('\n'.join(r))
          
def plot_scores(hits, misses):
    scores = pd.DataFrame(columns = ["score"], data = [ d[0] for d in (hits + misses) ])
    freq = scores.score.value_counts(sort=False).sort_index()
    freq.plot(kind="bar", figsize=(12,5), color=np.where(freq.index <= misses[0][0], 'blue', 'gold'))
```

## Loading pairings
Simple Pandas read_csv

The magic comes from .set_index which allows us to query the DataFrame by dbname for its ID (or ID for showname)

```python
# Load pairings
def load_pairings(id_file = "./dbname_2_id.csv", showname_file = "./db_id_2_showname.csv"):
    db_dbname_2_id = pd.read_csv(id_file)
    db_dbname_2_id = db_dbname_2_id.set_index("dbname") # Now we can do csv_data.loc[str(x)].id to get the id
    
    db_id_2_showname = pd.read_csv(showname_file)
    db_id_2_showname = db_id_2_showname.set_index("id")
    
    return db_dbname_2_id, db_id_2_showname
```

# Mapping functions
These functions expect that a relevent .set_index() has previously been called on the DataFrames (see load_pairings() and fuzzy_match() for example)
## db_map(index, df)
Mapping for dbname --> id or id --> showname

Give the index (dbname or id) and the corresponding DataFrame (db_dbname_2_id or db_id_2_showname) to obtain the mapping

None is returned if no pairing could be found. This must be taking into consideration by any calling code. (or else your DataFrame columns will become floats!)

## apply_add_id_column(db_dbname_2_id, sql_results, name_column, drop_null)
This will augment/append an ID column onto some sql_results

name_column is the name of the column in the sql_results that has the dbname aka candidate name

drop_null will drop any rows where the name_column column is not null. This is useful because if there is any null/None then all IDs become floats.

```python
# Mapping fucntions 
def db_map(index, df):
    try:
        return int(df.loc[index][0])
    except:
        return None # Consider changing this to -1. Any None causes id's to become floats in a DataFrame
    
def apply_add_id_column(db_dbname_2_id, sql_results, name_column, drop_null = False):
    def dbname_to_id(name):
        return db_map(name, db_dbname_2_id)
    
    if drop_null:
        # TODO This assignment causes a SettingWithCopyWarning
        sql_results = sql_results.dropna(subset=[name_column]) # Drop rows whos dbname is null, otherwise id becomes a float
        # sql_results = sql_results.loc[pd.notnull(sql_results[name_column]), :] # Drop rows whos dbname is null, otherwise id becomes a float
        
    sql_results.loc[:, "id"] = sql_results[name_column].apply(dbname_to_id)
    
    return sql_results
```
